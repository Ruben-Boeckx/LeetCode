```python
from typing import List
```

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        assert 1 <= len(strs) <= 200
        assert all(0 <= len(s) <= 200 for s in strs)
        prefix = ""

        for chars in zip(*strs):
            if len(set(chars)) == 1:
                prefix += chars[0]
            else:
                break

        return prefix
```

```python
strs = ["b"]

solution = Solution()
solution.longestCommonPrefix(strs)
```
