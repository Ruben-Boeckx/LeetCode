```python
import pandas as pd
import numpy as np
from typing import List
```

```python
class Solution:
    def __init__(self):
        pass

    def twoSum(self, nums: List[int], target: int) -> List[int]:
        assert 2 <= len(nums) <= 10**4
        assert -10**9 <= target <= 10**9

        for i, j in enumerate(nums):
            assert -10**9 <= j <= 10**9
            for k, l in enumerate(nums[i+1:]):
                if j + l == target and i != k + i + 1:
                    return [i, k + i + 1]
```

```python
nums = [3,3]
target = 6

solution = Solution()
result = solution.twoSum(nums, target)
print(result)
```
