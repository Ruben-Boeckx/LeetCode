```python
class Solution:
    def __init__(self):
        pass

    def threeSum(self, nums: List[int]) -> List[List[int]]:
        unique_results = list()
        assert 3 <= len(nums) <= 3000
        assert all(-10**5 <= num <= 10**5 for num in nums)

        for i, j in enumerate(nums):
            for k, l in enumerate(nums[i+1:]):
                for m, n in enumerate(nums[i+k+2:]):
                    if j + l + n == 0 and i != k + i + 1 and i != m + k + i + 2 and k + i + 1 != m + k + i + 2:
                        triplet = sorted([j, l, n])
                        
                        if (triplet[0], triplet[1], triplet[2]) not in unique_results:
                            unique_results.append((triplet[0], triplet[1], triplet[2]))

        return [list(triplet) for triplet in unique_results]
```

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        result = []

        n = len(nums)
        assert n >= 3 and n <= 3000

        for i in range(n):
            assert nums[i] >= -10**5 and nums[i] <= 10**5

            if i > 0 and nums[i] == nums[i - 1]:
                continue

            left, right = i + 1, n - 1

            while left < right:
                total = nums[i] + nums[left] + nums[right]

                if total == 0:
                    result.append([nums[i], nums[left], nums[right]])

                    left += 1
                    right -= 1

                    while left < right and nums[left] == nums[left - 1]:
                        left += 1
                    while left < right and nums[right] == nums[right + 1]:
                        right -= 1

                elif total < 0:
                    left += 1
                else:
                    right -= 1

        return result
```
