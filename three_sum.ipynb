```python
from typing import List
```

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        result = []

        n = len(nums)
        assert n >= 3 and n <= 3000

        for i in range(n):
            assert nums[i] >= -10**5 and nums[i] <= 10**5

            if i > 0 and nums[i] == nums[i - 1]:
                continue

            left, right = i + 1, n - 1

            while left < right:
                total = nums[i] + nums[left] + nums[right]

                if total == 0:
                    result.append([nums[i], nums[left], nums[right]])

                    left += 1
                    right -= 1

                    while left < right and nums[left] == nums[left - 1]:
                        left += 1
                    while left < right and nums[right] == nums[right + 1]:
                        right -= 1

                elif total < 0:
                    left += 1
                else:
                    right -= 1

        return result
```

```python
# nums = [-1,0,1,2,-1,-4]
# nums = [0,0,0]
nums = [0,1,1]

solution = Solution()
result = solution.threeSum(nums)
print(result)
```
