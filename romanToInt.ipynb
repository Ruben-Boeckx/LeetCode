```python
import re
```

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        if not (1 <= len(s) <= 15):
            raise ValueError("The string must contain between 1 and 15 characters.")

        if not re.fullmatch(r"^M{0,3}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$", s):
            raise ValueError("Invalid Roman numeral structure.")
        
        roman_map = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }

        if not all(char in roman_map for char in s):
            raise ValueError("Invalid characters found in the input string.")
        
        total = 0
        prev_value = 0

        for char in reversed(s):
            value = roman_map[char]
            print(f"Processing {char}: value={value}, prev_value={prev_value}, total={total}")
            
            if value < prev_value:
                total -= value
            else:
                total += value
            
            prev_value = value

        return total
```

```python
solution = Solution()
solution.romanToInt(s="MMMII")
```
